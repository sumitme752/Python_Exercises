1.What is the difference between population and sample?

A population is the entire group that you want to draw conclusions about. A sample is the specific group that you will collect data from. The size of the sample is always less than the total size of the population.



2.Explain the Central Limit Theorem.

The central limit theorem says that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough. Regardless of whether the population has a normal, Poisson, binomial, or any other distribution, the sampling distribution of the mean will be normal.

A normal distribution is a symmetrical, bell-shaped distribution, with increasingly fewer observations the further from the center of the distribution.

Central limit theorem formula
Fortunately, you don’t need to actually repeatedly sample a population to know the shape of the sampling distribution. The parameters of the sampling distribution of the mean are determined by the parameters of the population:

The mean of the sampling distribution is the mean of the population.
    \begin{equation*}\mu_{\bar{x}}=\mu\end{equation*}

The standard deviation of the sampling distribution is the standard deviation of the population divided by the square root of the sample size.
    \begin{equation*}\sigma_{\bar{x}} = \dfrac{\sigma}{\sqrt{n}}\end{equation*}

We can describe the sampling distribution of the mean using this notation:

    \begin{equation*}\bar{X}\sim N (\mu,\dfrac{\sigma}{\sqrt{n}})\end{equation*}

Where:

X̄ is the sampling distribution of the sample means
~ means “follows the distribution”
N is the normal distribution
µ is the mean of the population
σ is the standard deviation of the population
n is the sample size

3.What is p-value and how is it used in hypothesis testing?

The p-value is a probability value that measures the strength of evidence against the null hypothesis in hypothesis testing. It helps determine whether the observed data is statistically significant or simply due to random chance. A p-value less than a pre-determined significance level (often 0.05) suggests strong evidence against the null hypothesis.




4.Describe the concept of standard deviation.

Standard deviation is a measure of the variability or spread of a dataset. It quantifies how much the individual data points deviate from the mean of the dataset. A higher standard deviation indicates greater variability in the data.




5.What is the difference between correlation and covariance?

Correlation measures the linear relationship between two variables and ranges from -1 to +1, where -1 indicates a perfect negative correlation, +1 indicates a perfect positive correlation, and 0 indicates no correlation. Covariance, on the other hand, measures the direction and magnitude of the linear relationship between two variables, but it does not have a standardized scale like correlation.





6.Explain the difference between Type I and Type II errors.

Type I error, also known as a false positive, occurs when the null hypothesis is rejected when it is actually true. Type II error, also known as a false negative, occurs when the null hypothesis is accepted when it is actually false.




7.How would you handle missing data in a dataset?

Handling missing data can be done through techniques such as deletion (removing rows or columns with missing data), imputation (replacing missing values with estimated values), or using advanced methods like multiple imputation or expectation-maximization algorithms.




8.What is the purpose of confidence intervals?

Confidence intervals provide a range of values within which we can be reasonably confident that the population parameter lies. They help estimate the precision and uncertainty associated with sample statistics.



9.What is the difference between parametric and non-parametric statistics?

Parametric statistics assume a specific probability distribution for the data, such as normal distribution, and make inferences based on parameters. Non-parametric statistics make fewer assumptions about the underlying data distribution and use rank-based or distribution-free methods.




10.Describe the concept of statistical power.

Statistical power is the probability of correctly rejecting the null hypothesis when it is false. It depends on factors such as sample size, effect size, significance level, and variability in the data. A higher statistical power indicates a higher likelihood of detecting a true effect.




11.How would you determine if a given dataset is normally distributed?

Several methods can be used, including graphical methods like histogram or Q-Q plot, statistical tests like the Shapiro-Wilk test or Anderson-Darling test, or visual inspection of skewness and kurtosis measures.




12.What is the purpose of hypothesis testing and what are its steps?

Hypothesis testing is used to make inferences about population parameters based on sample data. The steps include stating the null and alternative hypotheses, choosing a significance level, calculating the test statistic, determining the critical region or p-value, and making a decision to reject or fail to reject the null hypothesis.




13.Explain the concept of bias-variance tradeoff.

The bias-variance tradeoff refers to the relationship between model complexity and its ability to generalize to unseen data. A high-bias model is simple and may underfit the data, while a high-variance model is overly complex and may overfit the data. Achieving a balance between bias and variance is crucial for building models with good predictive performance.



14.What is linear regression and how does it work?

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It assumes a linear relationship and estimates the coefficients of the linear equation using the method of least squares.



15.How would you interpret a p-value of 0.05?

A p-value of 0.05 means that there is a 5% chance of observing the obtained test statistic (or more extreme) under the assumption that the null hypothesis is true. If the p-value is less than the chosen significance level (usually 0.05), it suggests that the observed data is statistically significant, and we reject the null hypothesis.



16.What is the purpose of A/B testing and how would you design an A/B test?

A/B testing is used to compare two versions of a webpage, feature, or marketing campaign to determine which one performs better. To design an A/B test, you would identify the metric to measure, divide users into two groups (control and experimental), apply the changes to the experimental group, collect data, and analyze the results using statistical tests to determine the significance of the differences observed.



17.Explain the concept of outlier detection and how it can be done statistically.

Outlier detection is the process of identifying extreme or unusual observations in a dataset. Statistical techniques such as z-scores, box plots, or the interquartile range (IQR) can be used to identify outliers based on their deviation from the mean or quartiles.



18.What is the difference between supervised and unsupervised learning?

Supervised learning involves training a model with labeled data, where the desired outcome is known, to make predictions or classify new data. Unsupervised learning, on the other hand, deals with unlabeled data, and the model discovers patterns or structures in the data without specific target values.



19.Describe the concept of feature selection and its importance.

Feature selection involves identifying the most relevant and informative features from a dataset for building predictive models. It helps reduce dimensionality, improves model interpretability, reduces overfitting, and enhances model performance by focusing on the most important features.



20.How would you assess multicollinearity in a regression model?

Multicollinearity refers to the presence of high correlation between predictor variables in a regression model. It can be assessed using correlation matrices, variance inflation factor (VIF), or condition number. High values indicate high multicollinearity, which can affect the interpretability and stability of regression coefficients.

Hypothesis testing is a statistical procedure used to make inferences about population parameters based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H₀) and the alternative hypothesis (H₁), and assessing the evidence from the data to determine which hypothesis is more likely.

The null hypothesis represents the status quo or the default assumption, while the alternative hypothesis represents the claim or the hypothesis we want to test. The goal of hypothesis testing is to evaluate whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

The general process of hypothesis testing involves the following steps:

State the null and alternative hypotheses:

The null hypothesis (H₀) usually represents no effect, no difference, or no relationship between variables.
The alternative hypothesis (H₁) represents the opposite of the null hypothesis, stating that there is an effect, a difference, or a relationship.
Choose a significance level (α):

The significance level (α) determines the threshold for determining statistical significance. It is the probability of rejecting the null hypothesis when it is true. Commonly used significance levels are 0.05 (5%) and 0.01 (1%).
Select an appropriate test statistic:

The choice of test statistic depends on the nature of the data and the hypothesis being tested. Common test statistics include t-tests, chi-square tests, ANOVA, correlation coefficients, and regression coefficients.
Determine the critical region or calculate the p-value:

The critical region is the range of values for the test statistic that leads to the rejection of the null hypothesis. It is determined based on the chosen significance level.
The p-value is the probability of obtaining a test statistic as extreme as or more extreme than the one observed, assuming the null hypothesis is true. A smaller p-value indicates stronger evidence against the null hypothesis.
Compare the test statistic to the critical region or evaluate the p-value:

If the test statistic falls within the critical region, the null hypothesis is rejected in favor of the alternative hypothesis.
If the p-value is less than the significance level, the null hypothesis is rejected. Otherwise, the null hypothesis is not rejected.
Draw conclusions and make inferences:

Based on the results, conclusions are drawn regarding the evidence in favor of the alternative hypothesis or lack thereof. This helps make inferences about the population parameter being tested.
Hypothesis testing is a fundamental tool in statistics and is used to provide evidence for or against specific claims or hypotheses, aiding decision-making processes in various fields such as medicine, social sciences, business, and more.

T-Test:
A t-test is a statistical test used to compare the means of two groups and determine if they are significantly different from each other. It is commonly used when the sample size is small and the population standard deviation is unknown. The t-test calculates a test statistic called t-value and compares it to a critical value to determine statistical significance.

Z-Test:
A z-test is a statistical test used to compare means or proportions when the sample size is large, and the population standard deviation is known or assumed to be known. It is based on the standard normal distribution (z-distribution) and calculates a z-value as the test statistic. The z-value is then compared to a critical value to assess statistical significance.

Chi-Square Test:
A chi-square test is a statistical test used to determine if there is a significant association between categorical variables. It is often used to analyze the frequency distribution of categorical data. The chi-square test calculates a test statistic called chi-square (χ²) and compares it to a critical value from the chi-square distribution to determine statistical significance.

ANOVA (Analysis of Variance) Test:
ANOVA is a statistical test used to compare means across multiple groups. It determines whether there are significant differences in the means of three or more groups. ANOVA partitions the total variability in the data into two components: between-group variability and within-group variability. It calculates an F-statistic as the test statistic, which is then compared to a critical value from the F-distribution to assess statistical significance.

In summary:

T-tests are used to compare means between two groups.
Z-tests are used to compare means or proportions when the sample size is large.
Chi-square tests are used to analyze categorical data and determine association between variables.
ANOVA tests are used to compare means across multiple groups.
These tests are widely used in statistics and play important roles in hypothesis testing and data analysis. The choice of test depends on the nature of the data and the research question being investigated.

The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the sampling distribution of the mean (or sum) of a large number of independent and identically distributed random variables, regardless of the shape of the population distribution. It states the following:

For a random sample of n observations drawn from any population with a finite mean (μ) and finite standard deviation (σ), as n approaches infinity, the sampling distribution of the sample mean approaches a normal distribution.
The mean of the sampling distribution of the sample mean is equal to the population mean (μ).
The standard deviation of the sampling distribution of the sample mean, also known as the standard error (SE), is equal to the population standard deviation (σ) divided by the square root of the sample size (n).
In simpler terms, the Central Limit Theorem states that when we take repeated samples from any population, as the sample size increases, the distribution of sample means tends to be normal, regardless of the shape of the population distribution. This means that even if the population distribution is not normally distributed, the sample means will tend to follow a bell-shaped curve.

The Central Limit Theorem has several practical implications and applications:

It enables us to make inferences about population parameters based on sample means, assuming the sample is sufficiently large.
It justifies the use of many statistical techniques that rely on the assumption of normality, such as hypothesis testing and confidence intervals.
It helps explain why the normal distribution is commonly observed in practice, as many natural phenomena are the result of the sum of many small, independent, and random factors.
Overall, the Central Limit Theorem is a powerful tool in statistics that allows us to make probabilistic statements about population parameters based on sample data, providing a foundation for many statistical analyses and inference procedures.
